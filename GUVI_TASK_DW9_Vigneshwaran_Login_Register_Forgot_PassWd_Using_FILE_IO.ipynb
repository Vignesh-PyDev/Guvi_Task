```python
import re

def email_Validation():
    
    email = input("Enter Your Email Address:     ").lower().strip()
    
    if ('@' in email) and ("." in email):
        

        
        if email.index('@') < email.index('.'):
            
            
            if email[0].isalpha():
                
                Test_Str = [True if e.isalpha() else False for e in email[:email.index('@')] if e.isalpha()]
                
                if all(Test_Str):
                    

                    
                    if len(email[email.index('@'):email.index('.') + 1]) > 2:
            
                        print("\nEmail Validated\n")
                        

                        
                        return email
    

                    
                    elif email[email.index("@") + 1] == '.':
                        
                        print("No Immediate . is allowed after @\n")

                        return email_Validation()
                else:
                    
                    print("Only Alpha Numeric Allowed as a UserName\n")
                    
                    return email_Validation()
            else:
                
                print("No Special Characters Allowed at the Beginning\n")
                
                print(f'Your Email starts with {email[0]}\n\n\n')
                
                return email_Validation()
                
    else:
        
        print("Missing Mail Server Address or Domain\n")
        
        return email_Validation()
```

```python
def passWd_Validation():
    
    

    
    passwd = input("Enter Your Password:     ")
    

    
    if (len(passwd) > 5) and (len(passwd) < 16):
        

        
        if re.search(r'[\W]+',passwd):
            

            
            if re.search(r'[\d]+',passwd):
                

                
                if re.search(r'[A-Z]+',passwd):
                    

                    
                    if re.search(r'[a-z]+',passwd):
            
                        return passwd
                    
                    else:
                        print("Your Password Must Contain Minimum one Lower Case Character")
                            
                        return passWd_Validation()
                    
                else:
                    
                    print("Your Password Must Contain Minimum one Upper Case Character")
                    
                    return passWd_Validation()
                
                
                
            else:
                
                print("Your Password Must Contain Minimum one Numeric Character")
                
                return passWd_Validation()
            
        
            
        else:
            
            print("Your Password Must Contain Minimum one Special Character i.e @,#,$,% .,etc")
            
            return passWd_Validation()
        
    else:
        
        print("Your Password Length Should be Between 5 to 16 Characters Long")
        
        return passWd_Validation()
```
